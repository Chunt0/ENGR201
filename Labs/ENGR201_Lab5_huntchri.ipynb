## Measurements
-------------

Resistor Measurements:

$$R_1 = 986 \Omega \qquad R_2 = 670 \Omega \qquad R_3 = 462 \Omega \qquad R_{Load} = 2944 \Omega$$

Open Circuit Voltage, Short Circuit Current, Measured Total Resistance (Power Supply set to 0v):

$$V_{oc} = 3.65 V \qquad I_{sc}=4.19 mA \qquad R_{measured}=903 \Omega$$

$R_{Th}$ calculated from $V_{oc}$ and $I_{sc}$:
$$R_{Th} = \frac{V_{oc}}{I_{sc}} \Rightarrow R_{Th} = 871.12 \Omega$$

Current and Voltage Across $R_{Load}$:
$$I_L = 4.16 mA \qquad V_L = 2.814 v$$

Current and Voltage Across $R_{Load}$ with the Thevenin Equivalent Circuit:
$$I_L = 4.12 mA \qquad V_L = 2.809 v$$


## Theoretical
________________________
$V_{Th}$:


$$V_{Th} = 9v*\frac{670\Omega}{986\Omega+670\Omega} \Rightarrow V_{Th} = 3.64v$$

$R_{Th}$:


$$R_{Th} = \frac{986\Omega*670\Omega}{986\Omega+670\Omega}+462\Omega \Rightarrow R_{Th} = 865.93 \Omega$$

## Part 2
--------------



```python
import matplotlib.pyplot as plt

resistance = [100,200,350,500,650,800,871,950,1200,1500,2000,3000,5000,7000,10000]
power = [0.0014371681,0.00232562,0.00313801142857143,0.003553778,0.00375841384615385,0.00381938,0.00382390929965557,0.00381200947368421,0.0037276875,0.00357899266666667,0.0032131125,0.00264892033333333,0.0019269632,0.00149318414285714,0.0011162281]


plt.scatter(resistance, power)
plt.xlabel("Resistance, Ohms")
plt.ylabel("Power, Watts")
plt.title("Power Across Load Resistor")
plt.show()
```

As the load resistance increased the power across the load increased until a max was met. The max power occurred when $R_{Load}$ = 871. Power decreased as resistance increased past this resistance. This supports the theory that max power occurs at the Thevenin Resistance of the circuit.

Theoretical Max Power of the circuit:
$$P_{max} = \frac{V_{Th}^2}{4R_{Th}} \Rightarrow P_{max} = \frac{3.64^2}{4*865.93} \Rightarrow P_{max} = .00383W$$
Max Power of the circuit:
$$P_{max} = \frac{V_{Th}^2}{4R_{Th}} \Rightarrow P_{max} = \frac{1.825^2}{871} \Rightarrow P_{max} = .00382W$$
