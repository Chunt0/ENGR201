## Abstract
-----------------------
In this lab we investigate are to build two different circuits and to analyze them. First we will physically observe the circuits using a digital multimeter, then analytically, using node analysis. One circuit will be made entirely of resistors and another utilizing a diode as a fixed voltage source, this being done to simulate a "supernode".
